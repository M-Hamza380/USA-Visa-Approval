Store my data to MongoDB Atlas

```python
import os
import pymongo
```

```python
DB_NAME = 'Usa_Visa'
COLLECTION_NAME = 'usa_visa_data'
CONNECTION_URL = os.getenv('MONGODB_URL')
```

```python
client = pymongo.MongoClient(CONNECTION_URL)
data_base = client[DB_NAME]
collection = data_base[COLLECTION_NAME]
```

```python
record = collection.insert_many(data)
```

```python
records = collection.find()
records
```

```python
for i, j in enumerate(records):
    if i <= 10:
        print(f"{i}: {j}")
```
